```python
#barplot
result = sns.barplot(mtcars['cyl'], mtcars['mpg'])
plt.show()
```

```python
result = sns.barplot(x='cyl', y='mpg' ,data=mtcars)
plt.show()
```

```python
result = sns.barplot(x='cyl', y='mpg' ,data=mtcars, color= 'green')
plt.show()
```

```python
result = sns.barplot(mtcars['am'], mtcars['mpg'], palette = 'magma') #https://seaborn.pydata.org/tutorial/color_palettes.html
plt.show()
```
